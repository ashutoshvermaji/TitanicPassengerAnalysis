```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('passengerData.csv')
df
```

<font color='maroon' size=5>check null in survived and class both</font>

```python

for i in df.columns:
    print(f"number of null in values in {i}: {df[i].isnull().sum()}")
```

<font color='maroon' size=5>survived people according to class1 class2 and class3</font>

```python
grplist=[]
livelist=[]
diedlist=[]
grp=df.groupby(['Pclass','Survived'])
a=1
for g,data in grp:
    
    if g[0]==a and g[1]==0:
        grplist.append(f"class{g[0]}")
        diedlist.append(data['PassengerId'].count())
    elif g[0]==a and g[1]==1:
        livelist.append(data['PassengerId'].count())
        a+=1
        
        
print(livelist)
print(diedlist)
print(grplist)
    
```

```python
# visualization
fontstyle1={'family':'serif','color':'blue','size':25}
fontstyle2={'family':'serif','color':'maroon','size':15}
x=np.arange(len(grplist))
plt.bar(x-0.1,livelist,width=0.3,label='alive passenger')
plt.bar(x+0.2,diedlist,width=0.3,label='died passenger')
plt.xticks(x,grplist,fontdict=fontstyle2)
plt.ylabel('survived count',fontdict=fontstyle2)
plt.title('Passenger by class',fontdict=fontstyle1)
plt.legend()
plt.show()
```

<font color='maroon' size=5>survial age group vise</font>

```python
# create new column and store age group in it
df['AgeGroup']='Adult'
df['AgeGroup'].loc[df['Age']<18]='child'
df['AgeGroup'].loc[df['Age']>50]='old'
df['AgeGroup'].loc[df['Age'].isnull()]='others'
df
```

```python
# second method to create agegroup column 
def age(a):
    if a<18:
        return 'child'
    elif a>=18 and a<=50:
        return 'adult'
    elif a>50:
        return 'old'
    else:
        return 'others'
df['Agegroup']=df['Age'].map(age)
df
```

```python
# storing survival data according to age group in list
agegrp=['adult','child','old','others']
liveA=[]
diedA=[]
grp2=df.groupby(['Agegroup','Survived'])
a=0
for ag,data in grp2:
#     print(ag)
#     print(data)
    if ag[1]==0:
        diedA.append(data['PassengerId'].count())
    elif ag[1]==1:
        liveA.append(data['PassengerId'].count())
        a+=1
# print(liveA)
# print(diedA)


# visualization
fontstyle1={'family':'serif','color':'blue','size':25}
fontstyle2={'family':'serif','color':'maroon','size':15}
xA=np.arange(len(agegrp))
plt.bar(xA-0.1,liveA,label='Alive Passenger',width=0.3)
plt.bar(xA+0.2,diedA,label='Died Passenger',width=0.3)
plt.title('Passenger by age group',fontdict=fontstyle1)
plt.xticks(xA,agegrp,fontdict=fontstyle2)
plt.ylabel('Survived count',fontdict=fontstyle2)
plt.legend()
plt.show()
```

<font color='maroon' size=5>Gender vise Passenger survived information</font>

```python
sexlist=[]
liveS=[]
diedS=[]
grp3=df.groupby('Sex')
for sex,data in grp3:
    countL=0
    countD=0
#     print(sex)
#     print(data)
    sexlist.append(sex)
    for s in data['Survived']:
        if s==1:
            countL+=1
        elif s==0:
            countD+=1
    liveS.append(countL)
    diedS.append(countD)
print(sexlist)
print(liveS)
print(diedS)

# visualization
fontstyle1={'family':'serif','color':'blue','size':25}
fontstyle2={'family':'serif','color':'maroon','size':15}
x=np.arange(len(sexlist))
plt.bar(x-0.1,liveS,width=0.2,label='Alive Passenger')
plt.bar(x+0.1,diedS,width=0.2,label='Died Passenger')
plt.title('Survived by Gender',fontdict=fontstyle1)
plt.xticks(x,sexlist,fontdict=fontstyle2)
plt.ylabel('Survived count',fontdict=fontstyle2)
plt.legend()
plt.show()
    
```

<font color='maroon' size=5>survived by Embarked</font>

```python
# checking null values
print(f"number of null values in Embarked: {df['Embarked'].isnull().sum()}")
print(f"number of null values in survived: {df['Survived'].isnull().sum()}")
```

```python
# replacing null values
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
elist=[]
liveE=[]
diedE=[]
grp4=df.groupby('Embarked')
for e,data in grp4:
    elist.append(e)
    Dcount=0
    Lcount=0
#     print(e)
#     print(data)
    for s in data['Survived']:
        if s==0:
            Dcount+=1
        elif s==1:
            Lcount+=1
    liveE.append(Lcount)
    diedE.append(Dcount)
print(elist)
print(liveE)
print(diedE)

# visualization
fontstyle1={'family':'serif','color':'blue','size':25}
fontstyle2={'family':'serif','color':'maroon','size':15}
x=np.arange(len(elist))
plt.bar(x-0.1,liveE,width=0.3,label='Alive Passenger')
plt.bar(x+0.2,diedE,width=0.3,label='Died passenger')
plt.title('Passenger by Embarked',fontdict=fontstyle1)
plt.xticks(x,elist,fontdict=fontstyle2)
plt.ylabel('Survived count',fontdict=fontstyle2)
plt.legend()
plt.show()
```

<font color='maroon' size=6>survived by name title</font>

```python
# new column by extracting 'title' from name
df['NameTitle']=list(map(lambda name: name.split('.')[0].split(',')[-1].strip(),df['Name']))
# now replace value with 'others' except 'mr','mrs','miss','master'
df['NameTitle']=df['NameTitle'].map({'Mr':'Mr','Mrs':'Mrs','Miss':'Miss','Master':'Master'})
df['NameTitle'].fillna('Others',inplace=True)
```

```python
df['NameTitle'].unique()
```

```python
Ntitle=[]
liveT=[]
diedT=[]
grp5=df.groupby('NameTitle')
for t,data in grp5:
    Ntitle.append(t)
    diedcount=0
    livecount=0
    for i in data['Survived']:
        if i==0:
            diedcount+=1
        elif i==1:
            livecount+=1
    liveT.append(livecount)
    diedT.append(diedcount)
print(Ntitle)
print(liveT)
print(diedT)

# visualization
fontstyle1={'family':'serif','color':'blue','size':25}
fontstyle2={'family':'serif','color':'maroon','size':15}

x=np.arange(len(Ntitle))


plt.bar(x-0.2,liveT,width=0.4,label='Alive Passenger')
plt.bar(x+0.2,diedT,width=0.4,label='Died Passenger')
plt.title('Passenger by Name title',fontdict=fontstyle1)
plt.xticks(x,Ntitle,fontdict=fontstyle2)
plt.ylabel('Survived count',fontdict=fontstyle2)
plt.legend()
plt.show()
    
```
